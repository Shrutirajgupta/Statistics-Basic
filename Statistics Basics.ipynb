#Statistics Basics

1. Explain the different types of data (qualitative and quantitative) and provide examples of each. Discuss nominal, ordinal, interval, and ratio scales.

- > Qualitative Data: Describes non-numerical attributes or characteristics.

  Examples: Colors (red, blue), types of cars (SUV, sedan).

 Types:

 Nominal Scale: Categories without order (e.g., gender: male, female).

 Ordinal Scale: Categories with a meaningful order (e.g., satisfaction levels: low, medium, high).



- > Quantitative Data: Represents numerical values.

 Examples: Height, age, temperature.

 Types:

 Interval Scale: Numeric scale with no true zero (e.g., temperature in Celsius).

 Ratio Scale: Numeric scale with a true zero (e.g., weight, income).



2. What are the measures of central tendency, and when should you use each? Discuss the mean, median, and mode with examples and situations where each is appropriate.

 Mean: The average of data points.

 Use: When data is symmetrically distributed.

 Example: Average score of students in a class.


 Median: The middle value in an ordered dataset.

 Use: When data is skewed or has outliers.

 Example: Median income in a country.


 Mode: The most frequently occurring value.

 Use: For categorical data or identifying trends.

 Example: Most popular shoe size sold.


3. Explain the concept of dispersion. How do variance and standard deviation measure the spread of data?

 Dispersion: Describes how spread out data points are.

 Variance: Measures the average squared deviation from the mean.

 Standard Deviation: Square root of variance, representing dispersion in the same units as the data.

 Example: A small standard deviation implies data points are close to the mean.


4. What is a box plot, and what can it tell you about the distribution of data?

 Box Plot: A graphical representation of data distribution showing minimum, Q1, median, Q3, and maximum.

 Insights:

 Spread of data (IQR).

 Presence of outliers.

 Skewness based on the position of the median.


5. Discuss the role of random sampling in making inferences about populations.

 Random Sampling: A technique where each member of a population has an equal chance of selection.

 Role:

 Ensures unbiased representation.

 Facilitates valid statistical inferences.

 Example: Selecting 100 students randomly to estimate the average score of a university.



6. Explain the concept of skewness and its types. How does skewness affect the interpretation of data?

 Skewness: Describes the asymmetry in data distribution.

 Types:

 Positive Skew: Tail on the right; mean > median.

 Negative Skew: Tail on the left; mean < median.


 Effect: Influences the choice of central tendency (mean for symmetric, median for skewed).

9. Explain the properties of the normal distribution and the empirical rule (68-95-99.7 rule).

 Properties:

  Symmetrical and bell-shaped curve.

  Mean = Median = Mode.

  Defined by mean and standard deviation.


  Empirical Rule:

  68% of data lies within 1 SD.

  95% within 2 SDs.

  99.7% within 3 SDs.
